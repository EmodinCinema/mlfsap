Произведите разделение набора данных на три части: обучающая выборка (60%), тестовая выборка (10%), реальные данные (30%).

```python
import os
from tqdm import tqdm
import random

random.seed(425)
```

```python
images = os.listdir('/Users/User/jupyter_notebook/mlfsap/source/images/')
annotations = os.listdir('/Users/User/jupyter_notebook/mlfsap/source/annotations/')

print(images)
print(annotations)
```

```python
random.shuffle(images)

train_images = images[:round(len(images) * 0.6)]            # обучающая выборка (60%)
real_test_images = images[round(len(images) * 0.6):]        # тестовая выборка (10%) и реальные данные (30%)

real_images = real_test_images[:round(len(real_test_images) * 0.75)]        # реальные данные (30%)
test_images = real_test_images[round(len(real_test_images) * 0.75):]        # тестовая выборка (10%)

procent_train=(len(train_images)/(len(train_images)+len(real_test_images)))*100
procent_real=(len(real_images)/(len(train_images)+len(real_test_images)))*100
procent_test=(len(test_images)/(len(train_images)+len(real_test_images)))*100

print(train_images)
print(real_images)
print(test_images)
print('Изображения микросхем (images):', 
                                '\n', 'train -', len(train_images), f'{round(procent_train)}%',
                                '\n', 'real -', len(real_images), f'{round(procent_real)}%',  
                                '\n', 'test -', len(test_images), f'{round(procent_test)}%')
```

```python
train_path = '/Users/User/jupyter_notebook/mlfsap/dataset/image/train/'
test_path = train_path.replace('train', 'test')
val_path = train_path.replace('train', 'val')

source_path = '/Users/User/jupyter_notebook/mlfsap/source/images/'

print(train_path,'\n', test_path, '\n', val_path, '\n', source_path)
```

```python
annotations_name = train_images[0].replace(train_images[0].split('.')[-1], 'txt')
annotations_name
```

```python
def move_file(img_name: str, out_path):
    annotation_name = img_name.replace(img_name.split('.')[-1], 'txt')

    input_img_path = os.path.join(source_path, img_name)
    input_ann_path = os.path.join(source_path.replace('images', 'annotations'), annotation_name)

    if not os.path.exists(input_img_path):
        print(input_img_path , 'not_exist')
        return
    
    if not os.path.exists(input_ann_path):
        print(input_ann_path , 'not_exist')
        return

    os.rename(input_img_path, os.path.join(out_path, img_name))
    os.rename(input_ann_path, os.path.join(out_path.replace('image', 'labels'), annotation_name))
```

```python

for file in tqdm(train_images):
        move_file(file, train_path)
for file in tqdm(test_images):
        move_file(file, test_path)
for file in tqdm(real_images):
        move_file(file, val_path)
```
